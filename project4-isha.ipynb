```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
f = open("random.txt", "r")
newdf = []
list = []
for x in f:
  newlist =[]
  splitter = x.split()
  list.append(splitter[0])
  for itr in splitter[1:]:
    newlist.append(float(itr))
  data = {
    splitter[0] : newlist
  }
  newdf.append(data) 
print(newdf)
```

```python
newval = pd.DataFrame()
newval['bitap'] = newdf[0]['bitap']
newval['brute'] = newdf[3]['brute']
newval['rk'] = newdf[1]['rk']
newval['kmp'] = newdf[4]['kmp']
newval['bmh'] = newdf[2]['bmh']
indexa = [2,3,4,5,6,7,8,9,10]
newval['length of p'] = indexa
newval = newval.set_index('length of p')
print(newval)
```

```python
newval.plot.line()
```
